```python
import pandas as pd
import pickle
df= pd.read_csv('laliga.csv')
```

```python
df.head()
```

```python
#Data_cleaning
columns_to_remove=['Season','FTR','HTR']
final1=df.drop(labels=columns_to_remove,axis=1,inplace=True)
```

```python
df.head()
```

```python
teams=df['HomeTeam'].unique()
teams.sort()
teams
```

```python
key_team=['Alaves', 'Ath Bilbao', 'Ath Madrid',
       'Barcelona', 'Betis','Celta',
       'Eibar', 'Espanol', 'Getafe',
       'Granada','Leganes', 'Levante',
       'Mallorca','Osasuna',
       'Real Madrid', 'Sevilla',
       'Sociedad', 'Valencia', 'Valladolid', 'Villareal']
```

```python
df=df[(df['HomeTeam'].isin(key_team)) & (df['AwayTeam'].isin(key_team))]
```

```python
df
```

```python
#date-time conversion
from datetime import datetime
df['Date']=df['Date'].apply(lambda x:datetime.strptime(x,'%d-%m-%Y'))
```

```python
#data_processing
encoded_df= pd.get_dummies(data=df,columns=['HomeTeam','AwayTeam'])
```

```python
encoded_df
```

```python
encoded_df.columns
```

```python
req_df=encoded_df[['Date','HomeTeam_Alaves', 'HomeTeam_Ath Bilbao',
       'HomeTeam_Ath Madrid', 'HomeTeam_Barcelona', 'HomeTeam_Betis',
       'HomeTeam_Celta', 'HomeTeam_Eibar', 'HomeTeam_Espanol',
       'HomeTeam_Getafe', 'HomeTeam_Granada', 'HomeTeam_Leganes',
       'HomeTeam_Levante', 'HomeTeam_Mallorca', 'HomeTeam_Osasuna',
       'HomeTeam_Real Madrid', 'HomeTeam_Sevilla', 'HomeTeam_Sociedad',
       'HomeTeam_Valencia', 'HomeTeam_Valladolid', 'HomeTeam_Villareal',
       'AwayTeam_Alaves', 'AwayTeam_Ath Bilbao', 'AwayTeam_Ath Madrid',
       'AwayTeam_Barcelona', 'AwayTeam_Betis', 'AwayTeam_Celta',
       'AwayTeam_Eibar', 'AwayTeam_Espanol', 'AwayTeam_Getafe',
       'AwayTeam_Granada', 'AwayTeam_Leganes', 'AwayTeam_Levante',
       'AwayTeam_Mallorca', 'AwayTeam_Osasuna', 'AwayTeam_Real Madrid',
       'AwayTeam_Sevilla', 'AwayTeam_Sociedad', 'AwayTeam_Valencia',
       'AwayTeam_Valladolid', 'AwayTeam_Villareal','HTHG', 'HTAG','FTHG', 'FTAG']]
```

```python
X_train= req_df.drop(labels=['FTHG','FTAG'],axis=1)[req_df['Date'].dt.year>=2005]
X_test= req_df.drop(labels=['FTHG','FTAG'],axis=1)[req_df['Date'].dt.year>=2018]
```

```python
yh_train=req_df[req_df['Date'].dt.year>=2005]['FTHG'].values
yh_test=req_df[req_df['Date'].dt.year>=2018]['FTHG'].values
ya_train=req_df[req_df['Date'].dt.year>=2005]['FTAG'].values
ya_test=req_df[req_df['Date'].dt.year>=2018]['FTAG'].values
```

```python
X_train.drop(labels='Date',axis=True,inplace=True)
X_test.drop(labels='Date',axis=True,inplace=True)
```

# Regression

```python
#Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge1=Ridge()
parameters={'alpha':[1e-10,1e-8,1e-6,1e-4,1e-2,1,5,10,15,20,30,40,50]}
ridge1_regressor=GridSearchCV(ridge1,parameters,scoring='neg_mean_squared_error',cv=5)
ridge1_regressor.fit(X_train,yh_train)
```

```python
ridge2=Ridge()
parameters={'alpha':[1e-10,1e-8,1e-6,1e-4,1e-2,1,5,10,15,20,30,40,50]}
ridge2_regressor=GridSearchCV(ridge2,parameters,scoring='neg_mean_squared_error',cv=5)
ridge2_regressor.fit(X_train,ya_train)
```

```python
home='HomeTeam-Score-Model.pkl'
away='AwayTeam-Score-Model.pkl'
pickle.dump(ridge1_regressor,open(home,'wb'))
pickle.dump(ridge2_regressor,open(away,'wb'))
```
