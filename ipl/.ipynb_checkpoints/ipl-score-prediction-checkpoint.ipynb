```python
import pandas as pd
import pickle
```

```python
inn1=pd.read_csv('ipl_inn1.csv')
inn2=pd.read_csv('ipl_inn2.csv')
```

```python
encod_inn1=pd.get_dummies(data=inn1,columns=['bat_team','bowl_team'])
encod_inn2=pd.get_dummies(data=inn2,columns=['bat_team','bowl_team'])
```

```python
encod_inn1.columns
```

```python
req1=encod_inn1[['match','bat_team_CSK', 'bat_team_DC', 'bat_team_KKR', 'bat_team_KXIP',
       'bat_team_MI', 'bat_team_RCB', 'bat_team_RR', 'bat_team_SRH',
       'bowl_team_CSK', 'bowl_team_DC', 'bowl_team_KKR', 'bowl_team_KXIP',
       'bowl_team_MI', 'bowl_team_RCB', 'bowl_team_RR', 'bowl_team_SRH','over','runs','last_4','wickets','total']]
```

```python
req2=encod_inn2[['match','bat_team_CSK', 'bat_team_DC', 'bat_team_KKR', 'bat_team_KXIP',
       'bat_team_MI', 'bat_team_RCB', 'bat_team_RR', 'bat_team_SRH',
       'bowl_team_CSK', 'bowl_team_DC', 'bowl_team_KKR', 'bowl_team_KXIP',
       'bowl_team_MI', 'bowl_team_RCB', 'bowl_team_RR', 'bowl_team_SRH','over','runs','last_4','wickets','target','total']]
```

```python
req2
```

```python
inn1_x_train=req1.drop(labels=['total'],axis=1)[req1['match']<=450]
inn1_x_test=req1.drop(labels=['total'],axis=1)[req1['match']>450]
```

```python
inn2_x_train=req2.drop(labels=['total'],axis=1)[req2['match']<=450]
inn2_x_test=req2.drop(labels=['total'],axis=1)[req2['match']>450]
```

```python
inn1_y_train=req1[req1['match']<=450]['total'].values
inn1_y_test=req1[req1['match']>450]['total'].values
inn2_y_train=req2[req2['match']<=450]['total'].values
inn2_y_test=req2[req2['match']>450]['total'].values
```

```python
inn1_x_train.drop(labels='match',axis=True,inplace=True)
inn1_x_test.drop(labels='match',axis=True,inplace=True)
inn2_x_train.drop(labels='match',axis=True,inplace=True)
inn2_x_test.drop(labels='match',axis=True,inplace=True)
```

# Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
lasso1=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-5,1e-2,1,10,20,40,80,100,200,500,700]}
lasso1_regressor=GridSearchCV(lasso1,parameters,scoring='neg_mean_squared_error',cv=5)
lasso1_regressor.fit(inn1_x_train,inn1_y_train)
```

```python
print(lasso1_regressor.best_params_)
print(lasso1_regressor.best_score_)
```

```python
lasso2=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-5,1e-2,1,5,10,15,20,30,35,40]}
lasso2_regressor=GridSearchCV(lasso2,parameters,scoring='neg_mean_squared_error',cv=5)
lasso2_regressor.fit(inn2_x_train,inn2_y_train)
```

```python
pred1=lasso1_regressor.predict(inn1_x_test)
pred2=lasso2_regressor.predict(inn2_x_test)
```

```python
import seaborn as sns
```

```python
sns.distplot(inn1_y_test-pred1)
```

```python
sns.distplot(inn2_y_test-pred2)
```

```python
inning1='inn1-model-score.pkl'
inning2='inn2-model-score.pkl'
pickle.dump(lasso1_regressor,open(inning1,'wb'))
pickle.dump(lasso2_regressor,open(inning2,'wb'))
```
